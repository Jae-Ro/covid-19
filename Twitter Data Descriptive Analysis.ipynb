```python
tt = TweetTokenizer()
df_masks['full_text'] = df_masks['full_text'].apply(tt.tokenize)
df_masks.head()
```

```python
tweet_words = []
[tweet_words.extend(row['full_text']) for index, row in df_masks.iterrows()]
print(tweet_words)
```

```python
stop_words = list(set(stopwords.words('english')))
tweet_stopwords = ['RT', ':', '@', '‚Ä¶', '!', '.', '-', 'A', '‡πâ', ',', '‡πá', '‡∏ß', '‡∏π','‡∏µ', '‚Äô','‡∏∑', '‡πÉ‡∏™',
'‡πÅ‡∏ï','‡∏°','‡∏∏','‡∏ñ','‡∏≤','‡∏ó','pm2','‡∏°','@TrueCarnivorePK','‡∏™‡∏£','‡∏õ‡πÅ‡∏•','‡πÄ‡∏õ','precaution','mask‡πÄ‡∏Ç','‡∏¢‡∏ß‡∏Å','‡πÄ‡∏û','‡∏¢‡∏á‡∏û‡∏≠‡πÅ‡∏•','‡πÅ‡∏•‡∏∞‡∏•',
'‡∏≤‡∏á‡∏°','‡∏≠‡∏ö','‡∏≠‡∏¢‡πÜ','‡∏≤‡∏≠‡∏¢','5‡∏î','‡∏ñ','#','‡∏™‡∏£','‡πÄ‡∏õ','mask‡πÄ‡∏Ç','‡πÄ‡∏û','‡πÅ‡∏•‡∏∞‡∏•','‡∏≠‡∏ö','H','‡πÄ‡∏õ','‡πÄ‡∏û','‡∏≠‡∏ö','?','"','I','‡∏°','(', '[',
'Ô∏è',']','‡∏±','‚Äº',')']
stop_words.extend(tweet_stopwords)
print(stop_words)
```

```python
for word in tweet_words:
    if word in stop_words or'‡∏ô' in word:
        tweet_words.remove(word)
```

```python
word_frequencies = nltk.FreqDist(w for  w  in tweet_words)
most_common_words = word_frequencies.most_common(100)
for word in  most_common_words:
    print(word)
```

```python
links = []
for word in tweet_words:
    if 'https:' in word:
        links.append(word)
        
link_frequencies = nltk.FreqDist(link for  link  in links)
for url in link_frequencies.most_common(10):
    print(url)
```
